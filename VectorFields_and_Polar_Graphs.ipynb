Plot vector Field $\overrightarrow{F}= \textbf{i}-\textbf{j}$ 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = 1
v = -1

plt.quiver(x,y,u,v)
plt.show()
```

Plot $\overrightarrow{F}= \frac{x}{x^2+y^2}\textbf{i}+\frac{y}{x^2+y^2}\textbf{j}$ 

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = x/np.sqrt(x**2 + y**2)
v = y/np.sqrt(x**2 + y**2)

plt.quiver(x,y,u,v)
plt.show()
```
