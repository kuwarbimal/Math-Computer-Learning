Approximate the value of pi by taking the ratio of number of points in a unit square and number of points in the circle in first quadrant only using vectors.

```
def gen_rand_vecs(dims, number):
    bucket=[]
    """ generate uniformly distributed points in first quadrant for unit square."""
    vecs = np.random.uniform(low=0, high=1, size=(number,dims))
    for e in vecs:
      """All points in unit square"""
      plt.plot([0,e[0]], [0,e[1]], 'r')
      """All points in unit circle"""
      if e[0]**2+e[1]**2 <= 1:
        plt.plot([0,e[0]], [0,e[1]], 'b')
        bucket.append(e)
    plt.show()
  
    return (4*len(bucket))/number


c=rand_vecs = gen_rand_vecs(2, 10000)
c
```

```
def toeplitz_matrix(matr):
  m, n = matr.shape
  for row in range(m-1):
    for col in range(n-1):
      if matr[row, col] != matr[row+1, col+1]: return False
  
  return True

```

```
def tribonacci(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    elif n == 3: 
      return 1
    else:
        return tribonacci(n-1)+tribonacci(n-2)+tribonacci(n-3)
 
 
print(tribonacci(9))
```
